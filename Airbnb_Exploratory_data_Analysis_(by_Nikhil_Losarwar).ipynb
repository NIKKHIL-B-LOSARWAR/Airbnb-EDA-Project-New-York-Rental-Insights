Let's analyze the distribution and segmentation of listed rental properties on Airbnb across New York City. We will group the properties by the main regions, referred to as 'neighbourhood_group' (comprising 6 regions) in New York. Additionally, we'll derive several aggregated columns to gain important insights into these rental properties:

* Number of Neighborhood Areas : How many distinct neighborhoods (popular areas) exist within each main region of New York?
* Property Listings Count : How many properties are listed in each main region of New York City?
* Total Lifetime Reviews : What is the total number of reviews uploaded for properties in each region?
* Average Rental Price : What is the average price of rental properties in each region?
* Median Rental Price : What is the median price of rental properties in each region?
* Number of Shared rooms  : In each region, how many shared room type properties are listed?
* Number of Private rooms  : In each region, how many private room type properties are listed?
* Number of Entire houses/apartments: In each region, how many entire houses/apartment type properties are listed?



````python
  # Let's summarize and derive some aggregate values/columns using groupby and agg. functions

agg_df_1 = airbnb_df.groupby('neighbourhood_group').agg({
    'id': 'count',
    'host_id' : 'nunique',
    'neighbourhood' : 'nunique',
    'number_of_reviews': 'sum',
    'price' : ['mean', 'median']

}).reset_index()

# renaming the columns name
agg_df_1.columns = [
    'region_group',         # Original name for grouping column
    'property_count',       # Count of 'id'
    'Host_count',           # unique count of 'host_id'
    'areas_count',                 # Unique count of 'neighbourhood'
    'reviews_count',               # Sum of 'number_of_reviews'
    'avg_price',                   # Mean of 'price'
    'median_price'                 # Median of 'price'

]


print(agg_df_1)
````

````python
# Create a new DataFrame with counts of different room types after grouping by 'neighbourhood_group'
agg_df_2 = airbnb_df.groupby('neighbourhood_group').agg(
    shared_room_count = ('room_type', lambda x: (x == 'Shared room').sum()),
    private_room_count = ('room_type', lambda x: (x == 'Private room').sum()),
    entire_home_count = ('room_type', lambda x: (x == 'Entire home/apt').sum())
).reset_index()

agg_df_2.columns = [
    'region_group',
    'shared_room_count',
    'private_room_count',
    'entire_home_count'

]

# Display the resulting DataFrame
print(agg_df_2)
````

````python
# let's join/merge the two above dataframes
agg_df = pd.merge(agg_df_1, agg_df_2, on = 'region_group')

# let's print the final resultant dataframe
print(agg_df)
````

#### Chart - 7 : Line Chart - Reviews trend by Months

````python
# Chart - 7 visualization code

# Creating new dataframe for this visualization

# Convert 'last_review' column to datetime if it's not already
airbnb_df['last_review'] = pd.to_datetime(airbnb_df['last_review'], errors='coerce')

# Create the new dataframe with 'id', 'last_review_month', and 'last_review_year' columns
df_last_review_mth_yr = airbnb_df[['id']].copy()
df_last_review_mth_yr['last_review_month'] = airbnb_df['last_review'].dt.strftime('%B')
df_last_review_mth_yr['last_review_month_no'] = airbnb_df['last_review'].dt.month
df_last_review_mth_yr['last_review_year'] = airbnb_df['last_review'].dt.year

# Display the new dataframe
df_last_review_mth_yr.head()
````

````python

# Remove rows with any NaN or null values from new_df
df_last_review_mth_yr = df_last_review_mth_yr.dropna()

 # To check the count of nulls present in each column
print("Null count in each column => \n", df_last_review_mth_yr.isnull().sum())
````
